```python
#join/merge with same column name
import pandas as pd
```

```python
emp={"id":[101,102,103,104,105],
    "name":['a','b','c','d','e'],
     "city":["pune","mumbai","chennai","banglore","delhi"]
    }
```

```python
dept={"id":[102,104,555,190,105],
      "dname":["IT","HR","Admin","IT","Finance"],
       " sal":[55000,90000,62000,30000,20000]
}
```

```python
emp_df=pd.DataFrame(emp)
```

```python
dept_df=pd.DataFrame(dept)
```

```python
emp_df
```

```python
dept_df
```

```python
emp_df.merge(dept_df)
```

```python
emp_df.merge(dept_df,how='left')
```

```python
emp_df.merge(dept_df,how='right')
```

```python
#join/merge with diff column 
emp1={"empid":[101,102,103,104,105],
    "name":['a','b','c','d','e'],
     "city":["pune","mumbai","chennai","banglore","delhi"]
    }
```

```python
dept1={"did":[102,104,555,190,105],
      "dname":["IT","HR","Admin","IT","Finance"],
       " sal":[55000,90000,62000,30000,20000]
}
```

```python
emp1_df= pd.DataFrame(emp1)
```

```python
dept1_df= pd.DataFrame(dept1)
```

```python
emp1_df
```

```python
dept1_df
```

```python
emp1_df.merge(dept1_df, left_on ="empid" , right_on = "did")
```
